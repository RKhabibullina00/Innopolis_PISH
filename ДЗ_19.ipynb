```python
from sklearn.preprocessing import normalize
df3=df1[df1.columns[3:12]]
#df3
normalized_df=normalize(df3, norm='l2', axis=1)
normalized_df3=pd.DataFrame(normalized_df, columns=df3.columns)
normalized_df3
```

```python
percent_missing = df1.isnull().sum() * 100 / len(df1)
percent_missing
```

```python
df1=df.loc[:10]
df1['1990']=df1['1990'].fillna(method='bfill')
df1['1988']=df1['1988'].fillna(method='ffill')
df1['1986']=df1['1986'].fillna(df1['1986'].mean())
df1['1984']=df1['1984'].fillna(df1['1984'].median())
df1
```
